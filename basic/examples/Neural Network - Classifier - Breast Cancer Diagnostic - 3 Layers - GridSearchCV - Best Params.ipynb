### Importing data

```python
import pandas as pd
```

```python
features = pd.read_csv('datasets/breast_cancer_wisconsin_diagnostic_data.csv')
```

```python
classes = pd.read_csv('datasets/breast_cancer_wisconsin_diagnostic_classes.csv')
```

### Building the Neural Network with Dropout

The use of Dropout helps to prevent the occurrence of Overfitting, as it resets some neurons based on a specified percentage, giving the model greater variability in predictions.

Dropout: https://keras.io/api/layers/regularization_layers/dropout/

```python
import math
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
def build_model(input_units, activation, kernel_initializer, optimizer, loss):
    input_dim = features.shape[1]
    output_units = 1

    model = Sequential()
    model.add(Dense(units=input_units, activation=activation, kernel_initializer=kernel_initializer, input_dim=input_dim))
    model.add(Dropout(0.2))
    model.add(Dense(units=input_units, activation=activation, kernel_initializer=kernel_initializer))
    model.add(Dropout(0.2))
    model.add(Dense(units=output_units, activation='sigmoid'))
    
    model.compile(optimizer=optimizer, loss=loss, metrics=['binary_accuracy'])
    
    return model
```

### Applying GridSearchCV

GridSearchCV is used to identify the best set of parameters within a sequence.

These parameters will later be used to generate a model with the highest accuracy.

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
classifier = KerasClassifier(build_fn=build_model)
```

```python
params = {
    'input_units': [8, 16],
    'activation': ['relu', 'tanh'],
    'kernel_initializer': ['random_uniform', 'normal'],
    'optimizer': ['adam', 'sgd'],
    'loss': ['random_crossentropy', 'hinge'],
    'batch_size': [10, 30],
    'epochs': [50, 100]
}
```

```python
grid = GridSearchCV(estimator=classifier, param_grid=params, scoring='accuracy', cv=5)
```

```python
grid.fit(features, classes)
```

### Validating best params

```python
best_params = grid.best_params_
```

```python
print('Best params')
```

```python
for k, v in best_params.items():
    print(f'{k}: {v}')
```

```python
best_score = grid.best_score_
```

```python
print(f'Best score: {best_score}')
```
